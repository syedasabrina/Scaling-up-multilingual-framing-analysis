```python
import pandas as pd
import re
from sklearn.metrics import accuracy_score
```

# Calculate accuracy

```python
df = pd.read_csv('data/falcon_preds.csv', sep=',')
# df = pd.read_csv('data/mistral7b_preds_few_shot.csv', sep=',')
# df = pd.read_csv('data/llama_preds.csv', sep=',')
preds = list(df['prediction'])
```

```python
def clean_text(text):
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    # Remove special characters and punctuation
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    # Remove extra spaces and line breaks
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
preds = [str(text) for text in preds]
```

```python
cleaned_text_data = [clean_text(text) for text in preds]
df['clean_preds'] = cleaned_text_data
df
```

```python
actual_labels = list(df['output'])
predicted_labels = list(df['prediction'])
accuracy = accuracy_score(actual_labels, predicted_labels)
accuracy
```

# fasttext

```python
import fasttext
import numpy as np

# Load the pre-trained FastText model
model = fasttext.load_model("content/wiki.en.bin")
# model = fasttext.load_model("content/cc.en.300.bin")

def get_sentence_vector(sentence):
    """Get the averaged word vector representation of a sentence."""
    words = sentence.split()
    if not words:
        return np.zeros(model.get_dimension())
    return sum([model.get_word_vector(word) for word in words]) / len(words)

def closest_label(input_string, labels):
    """Find the label that is closest to the input_string."""
    input_vector = get_sentence_vector(input_string)
    similarities = [np.dot(input_vector, get_sentence_vector(label)) for label in labels]
    return labels[np.argmax(similarities)]

def string_match(input_string):
    labels = ["Economic", "Capacity and Resources", "Morality", "Fairness and Equality", "Legality, Constitutionality, Jurisdiction", 
              "Policy Prescription and Evaluation", "Crime and Punishment", "Security and Defense", "Health and Safety", 
              "Quality of Life", "Cultural Identity", "Public Sentiment", "Political", "External Regulation and Reputation", "Other"]
    result = closest_label(input_string, labels)
    return result

def accuracy_calculate(out, preds):
    actual_labels = list(out)
    predicted_labels = list(preds)
    accuracy = accuracy_score(actual_labels, predicted_labels)
    return accuracy
```

```python
# df = pd.read_csv('data/llama_predictions.csv', sep=',')
# df.columns = df.columns.str.strip()
preds = list(df['clean_preds'])
# preds = ['Legality, Constitutionality, Jurisdiction' if 'Legal' in str(s) else str(s) for s in preds]
preds2 = [string_match(s) for s in preds]
df['fasttext'] = preds2
accuracy_calculate(df['output'], df['fasttext'])
```

# edit distance

```python
import editdistance

def closest_label(input_string, labels):
    """Find the label that is closest to the input_string using the edit distance."""
    distances = [editdistance.eval(input_string, label) for label in labels]
    return labels[distances.index(min(distances))]

def string_match(input_string):
    labels = ["Economic", "Capacity and Resources", "Morality", "Fairness and Equality", "Legality, Constitutionality, Jurisdiction", 
              "Policy Prescription and Evaluation", "Crime and Punishment", "Security and Defense", "Health and Safety", 
              "Quality of Life", "Cultural Identity", "Public Sentiment", "Political", "External Regulation and Reputation", "Other"]
    result = closest_label(input_string, labels)
    return result

def accuracy_calculate(out, preds):
    actual_labels = list(out)
    predicted_labels = list(preds)
    accuracy = accuracy_score(actual_labels, predicted_labels)
    return accuracy
```

```python
# df = pd.read_csv('data/llama_predictions.csv', sep=',')
preds = list(df['clean_preds'])
preds = ['Legality, Constitutionality, Jurisdiction' if 'Legal' in str(s) else str(s) for s in preds]
preds2 = [string_match(s) for s in preds]
df['edit_distance'] = preds2
accuracy_calculate(df['output'], df['edit_distance'])
```

# GPT3.5

```python
from bert_score import score


def cal_bertscore(filename):
    df = pd.read_csv(filename, sep=',')

    # Extract the "output" and "prediction" columns as lists
    references = df['output'].tolist()
    predictions = df['prediction'].tolist()

    # Calculate BERTScore
    P, R, F1 = score(predictions, references, lang='en')

    # Calculate the mean BERTScore
    mean_P = P.mean().item()
    mean_R = R.mean().item()
    mean_F1 = F1.mean().item()

    print(f"Overall BERTScore: P={mean_P:.4f}, R={mean_R:.4f}, F1={mean_F1:.4f}")
```

```python
filename = 'data/llama_preds.csv'
cal_bertscore(filename)
```
